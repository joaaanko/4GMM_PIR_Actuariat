### Test de convergence de P(X_1 + X_2 > t) / P(X_1 > t) vers 2 

```python
def get_Int(bin_edges, t): # Retourne l'indice du premier bord de bin dont la réalisation est au dessus de t
    nb = len(bin_edges)
    i = 0
    done = 0
    if t>bin_edges[-1]:
        print("aucune réalisation suppérieure à t")
    if t <= bin_edges[-1]:
        while done == 0 :
            if bin_edges[i] < t: 
                i = i + 1
            if bin_edges[i] > t:
                done = 1 
    return i 
```

```python
def calc_Prob(t, hist, bin_edges): #Calcule P(X>t)
    return np.sum(hist[t:]*np.diff(bin_edges[t:]))
```

```python
N = 100000000
X_1 = np.random.pareto(2, size=N)
X_2 = np.random.pareto(2, size=N)
hist, bin_edges = np.histogram(X_1+X_2,bins=1000, density = True)
```

```python
X_3 = X_1 #np.random.pareto(2, size=N)
hist_1, bin_edges_1 = np.histogram(X_3,bins=1000, density = True)
```

```python
Tfin = 2000
```

```python
P = np.zeros(Tfin)
for t in range(1,Tfin,1):
    P_Somme = calc_Prob(get_Int(bin_edges,t), hist, bin_edges)
    P_X3 =  calc_Prob(get_Int(bin_edges_1,t), hist_1, bin_edges_1)
    P[t] = P_Somme/P_X3
    
```

```python
plt.plot(P)
plt.grid( alpha=0.75)
```

### Somme et max de 3 Pareto 

```python

X_a = np.random.pareto(2, size=N)
X_b = np.random.pareto(2, size=N)
X_c = np.random.pareto(2, size=N)
#print(np.maximum(X_a,X_b,X_c))
hist2, bin_edges2 = np.histogram(X_a+X_b+X_c,bins=1000, density = True)
hist3, bin_edges3 = np.histogram(np.maximum(X_a,X_b,X_c),bins=1000, density = True)
```

```python
alpha=2
P2 = np.zeros(Tfin)
P3 = np.zeros(Tfin)

for t in range(1,Tfin,1):
    P_Somme = calc_Prob(get_Int(bin_edges2,t), hist2, bin_edges2)
    Pmax=calc_Prob(get_Int(bin_edges3,t), hist3, bin_edges3)
    denom =  1/t**alpha
    P2[t] = P_Somme/denom
    P3[t] = Pmax/denom
```

```python
plt.plot(P2)
plt.grid( alpha=0.75)
plt.title ("Somme de 3 Pareto")
plt.show()
plt.plot(P3)
plt.grid( alpha=0.75)
plt.title("Max de 3 Pareto")
plt.show()
```
