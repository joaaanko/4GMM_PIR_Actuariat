```python
#loi de pareto
def conv_pareto (alpha):
    
    t=np.linspace(10,1000000, 100)
    proba=[]
    for i in t:
        denom=(1+i)**alpha
        numer= nume (alpha,i)
        proba.append (numer/denom)
    plt.plot (t,proba)
    return t,proba


conv_pareto(10)
```

```python
#loi de pareto

def conv_pareto2 (alpha):
    num= 1000
    t=np.linspace(10,10000, 100)
    proba=[]
    
    
    for i in t:
        denom=0
        nume=0
        for n in range (num):
            
            X1 =np.random.pareto(alpha)
            X2=np.random.pareto (alpha)
            nume+=1*(X1>=t)
            denom+=1*(X1+X2>=t)
        proba.append(nume/denom)
        
        
    return proba

probabilite = conv_pareto2 (5)
print(probabilite)
```

```python
#travail 2

#Avec la valeur absolue de la gaussienne

def determ_K_gauss (N, mu, sigma, lambd,c):
    K=np.linspace (50,60,50)
    res=[]
    for k in K :
        f=[]
        for i in range (N):
    
            X1=np.random.normal(mu,sigma)
            W1=np.random.exponential (lambd)
            f.append(np.exp(k*(np.abs(X1)-c*W1)))
        res.append((np.sum(f))/N)
        
    plt.plot(K,res)
    plt.plot(K,np.ones(len(K))) # valeur thorique
    plt.title("Recherche du K tq f(K)=1")
    plt.show()
    return (K,res)

        
#Avec l'uniforme sur a,b positifs

def determ_K_unif (N, lambd,c):
    K=np.linspace (50,70,100)
    res=[]
    for k in K :
        f=[]
        for i in range (N):
    
            X1=np.random.uniform (0,1)
            #print (X1)
            W1=np.random.exponential (lambd)
            #print (W1)
            f.append(np.exp(k*(X1-c*W1)))
        
        res.append((np.sum(f))/N)
        
    
    plt.plot(K,res)
    plt.plot(K,np.ones(len(K))) # valeur thorique
    plt.title("Recherche du K tq f(K)=1")
    plt.show()
    return (K,res)

#avec l'exponentiellle pour comparer avec le rsultat calcul (K=(c*lamb_tild-lamb)/c)

def determ_K_exp (N,lamb_t, lambd,c):
    K=np.linspace (0,10,100)
    res=[]
    for k in K :
        f=[]
        for i in range (N):
    
            X1=np.random.exponential (lamb_t)
            W1=np.random.exponential (lambd)
            f.append(np.exp(k*(X1-c*W1)))
        
        res.append((np.sum(f))/N)
        
    plt.plot(K,res)
    plt.plot(K,np.ones(len(K))) # valeur thorique
    plt.title("Recherche du K tq f(K)=1")
    plt.show()
    return (K,res)



determ_K_gauss (1000, 0,1,1,100)
determ_K_exp (1000,3,2,4)
```
